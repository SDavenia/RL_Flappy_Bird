```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def my_function(x):
    if x <= 200:
        return 0.2
    else:
        return 0.2 / (1 + 0.005 * (x - 200)**1.05)

# Generate x-values
x = np.linspace(0, 1000, 100000)

# Compute y-values using the function
y = np.array([my_function(xi) for xi in x])

# Plot the function
plt.plot(x, y)
plt.xlabel('x')
plt.ylim(0, 0.220)
plt.ylabel('y')
plt.title('')
plt.show()
```
